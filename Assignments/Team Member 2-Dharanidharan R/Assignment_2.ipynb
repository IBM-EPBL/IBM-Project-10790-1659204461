# 6 - Find Outliers and Replacing

```python
sns.boxplot(x=df["EstimatedSalary"])
```

```python
sns.boxplot(x=df["CreditScore"])
```

```python
## Performing Inter Quartile Range
q2 = df.CreditScore.quantile(0.25)
q3 = df.CreditScore.quantile(0.75)
IQR = q3 - q2
upper_limit = q3+1.5*IQR
lower_limit = q2-1.5*IQR
```

```python
print(lower_limit)
print(upper_limit)
df["CreditScore"].head
```

```python
df["CreditScore"] = np.where(df["CreditScore"]<lower_limit,6.52,df["CreditScore"])
```

```python
df["CreditScore"] = np.where(df["CreditScore"]<lower_limit,6.52,df["CreditScore"])
sns.violinplot(x=df["CreditScore"])
```

```python
sns.boxplot(x=df["CreditScore"])
```
